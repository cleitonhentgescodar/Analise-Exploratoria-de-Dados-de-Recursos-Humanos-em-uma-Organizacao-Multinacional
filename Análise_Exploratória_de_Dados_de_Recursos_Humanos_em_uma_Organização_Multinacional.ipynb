# HR Dataset (Multinational Company)

O conjunto de dados utilizado neste estudo refere-se a informações de **Recursos Humanos** de uma **organização multinacional fictícia**, contendo registros detalhados de colaboradores e suas respectivas características profissionais e demográficas.  

Esse dataset é amplamente utilizado em projetos de **Análise Exploratória de Dados (EDA)** voltados à compreensão de fatores relacionados à **retenção de talentos**, **desempenho** e **experiência dos colaboradores**.  

O conjunto é composto por aproximadamente **2.000.000 de registros**, abrangendo variáveis que descrevem o **perfil profissional**, o **histórico de carreira** e as **avaliações internas** dos funcionários.  

**Fonte:** [Kaggle – HR Dataset (Multinational Company)](https://www.kaggle.com/datasets/rohitgrewal/hr-data-mnc)

---

## Principais Variáveis do Conjunto de Dados

- **Employee_ID:** Identificador único atribuído a cada colaborador.  
- **Full_Name:** Nome completo do funcionário.  
- **Gender:** Gênero do colaborador (Masculino, Feminino, etc.).  
- **Department:** Departamento de atuação (TI, RH, Finanças, Operações, etc.).  
- **Job_Title:** Cargo ou função exercida pelo colaborador.  
- **Hire_Date:** Data de admissão do funcionário na empresa.  
- **Location:** Localização geográfica da unidade (cidade e país).  
- **Experience_Years:** Tempo total de experiência profissional do colaborador.  
- **Performance_Rating:** Avaliação de desempenho (escala numérica, quanto maior, melhor).  
- **Promotions:** Número de promoções recebidas ao longo da carreira.  
- **Status:** Situação atual do vínculo empregatício (Ativo, Resignado, etc.).  
- **Work_Mode:** Modalidade de trabalho (Presencial, Híbrido, Remoto).  
- **Salary_INR:** Salário anual do colaborador, em Rúpias indianas.  

---

## Relevância do Dataset

Este conjunto de dados possibilita a realização de **análises descritivas e inferenciais** voltadas à compreensão dos **padrões de desempenho, experiência e retenção** de colaboradores.  

Além disso, serve como uma **base de aprendizado prático** para o desenvolvimento de competências em **Ciência de Dados aplicada a Recursos Humanos**, permitindo o aprimoramento de habilidades em **análise, interpretação e visualização de dados corporativos**.



```python
# importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Carregando Dados
df = pd.read_csv('/content/HR_Data_MNC_Data Science Lovers.csv')
display(df)
```

# Limpeza e Pré-processamento de Dados

```python
# Verificando a presença de valores nulos e tipos de dados
print('Verificando a presença de valores nulos e tipos de dados')
df.info()

# Removendo a coluna 'Unnamed: 0' pois não sera utilizada.
df.drop( 'Unnamed: 0', axis = 1, inplace = True)
display(df)

# Alterando o tipo de dados da coluna 'Hire_Date' para datetime.
df['Hire_Date'] = pd.to_datetime( df['Hire_Date'] )

print('\n')

# Verificando alteração.
print('Verificando alteração')
df.info()
```

1) Qual é a distribuição do status do funcionário (ativo, demitido, aposentado, demitido)?

```python
# Criando um gráfico de pizza para visualizar a distribuição dos status dos funcionários.

# Contando valores na coluna 'Status'.
status = df['Status'].value_counts()

# Contagem de status.
display(status)

# Paleta de cores suaves
cores = ['#66c2a5', '#fc8d62', '#8da0cb', '#e78ac3']

# Criando o gráfico de pizza
fig, ax = plt.subplots(figsize=(6, 6), facecolor='white')

status.plot(
    kind='pie',
    colors=cores,
    autopct='%1.1f%%',
    explode=(0.03, 0.03, 0.03, 0.03),
    startangle=90,
    shadow=False,
    ax=ax
)

# Design
plt.title('Distribuição dos Status dos Funcionários', fontsize=14, fontweight='bold', pad=20)
plt.ylabel('')
plt.tight_layout()

# Removendo o contorno
for spine in ax.spines.values():
    spine.set_visible(False)

plt.show()
```

```python
# Criando um gráfico de pizza para visualizar a distribuição dos modos de trabalho.

# Paleta moderna (viridis)
cores = sns.color_palette('viridis', n_colors=len(work))

# Criando o gráfico de pizza
fig, ax = plt.subplots(figsize=(6, 6), facecolor='white')

# Gráfico de pizza com efeito "donut"
work.plot(
    kind='pie',
    colors=cores,
    autopct='%1.1f%%',
    startangle=90,
    explode=[0.03] * len(work),
    shadow=False,
    ax=ax
)

# Adiciona círculo branco no meio (efeito "donut")
centre_circle = plt.Circle((0, 0), 0.30, fc='white')
fig.gca().add_artist(centre_circle)

# Título e estilo
plt.title('Distribuição dos Modos de Trabalho', fontsize=14, fontweight='bold', pad=20)
plt.ylabel('')
ax.set_aspect('equal')
plt.tight_layout()

# Remove bordas do gráfico
for spine in ax.spines.values():
    spine.set_visible(False)

plt.show()
```

4) Qual é o salário médio por departamento?

```python
# Calculando a média de salário de cada departamento
deptr = df.groupby('Department')['Salary_INR'].mean()/1000
display(deptr)
```

```python
# Criando um gráfico de barras para visualizar o salário médio por departamento.
deptr_sorted = deptr.sort_values(ascending=False).reset_index()
deptr_sorted.columns = ['Departamento', 'Salário Médio']

# Estilo visual moderno
sns.set(style="whitegrid", context="talk")

# Criação do gráfico
plt.figure(figsize=(10, 6))
ax = sns.barplot(
    data=deptr_sorted,
    x='Departamento',
    y='Salário Médio',
    palette='viridis'
)

# Título e eixos
plt.title('Salário Médio por Departamento', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('Departamento', fontsize=13)
plt.ylabel('Salário Médio', fontsize=13)

# Rotação e legibilidade
plt.xticks(rotation=45, ha='right', fontsize=11)
plt.yticks(fontsize=11)

# Ajustes visuais
sns.despine(left=True, bottom=True)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()

plt.show()
```

7) Quantos funcionários pediram demissão e foram demitidos em cada departamento?

```python
# Criando um gráfico de barras para visualizar a contagem de funcionários demitidos e desligados por departamento.
# Combina as informações de demitidos e desligados
combined_status = pd.DataFrame({
    'Pediram Demissão': R_emp,
    'Foram Demitidos': T_emp
}).fillna(0)

# Ordena pelo total (soma das duas colunas)
combined_status['Total'] = combined_status.sum(axis=1)
combined_status = combined_status.sort_values('Total', ascending=False).drop(columns='Total')

# Estilo visual consistente
sns.set(style="whitegrid", context="talk")

# Cores suaves e harmoniosas
cores = ['#2ca02c', '#e74c3c']  # verde-azulado e vermelho suave

# Criação do gráfico
ax = combined_status.plot(
    kind='bar',
    figsize=(11, 6),
    color=cores,
    edgecolor='black',
    width=0.7
)

# Título e eixos
plt.title(' Funcionários que Pediram Demissão e Foram Demitidos por Departamento',
          fontsize=17, fontweight='bold', pad=20)
plt.xlabel('Departamento', fontsize=13)
plt.ylabel('Quantidade de Funcionários', fontsize=13)

# Rótulos do eixo X e Y
plt.xticks(rotation=45, ha='right', fontsize=11)
plt.yticks(fontsize=11)

# Legenda
plt.legend(title='Status do Funcionário', fontsize=11, title_fontsize=12)

# Grade e estilo
plt.grid(axis='y', linestyle='--', alpha=0.3)
sns.despine(left=True, bottom=True)

plt.tight_layout()
plt.show()
```

15) Identifique os departamentos com a maior taxa de rotatividade (% de demissões)

```python
# Agrupa o DataFrame por 'Department' e calcula o total de funcionários e funcionários que pediram demissão.
dept_counts = df.groupby('Department')['Status'].agg(
    total_emp = 'count',  # Conta o número total de funcionários por departamento.
    resigned = lambda x: (x == 'Resigned').sum() # Conta o número de funcionários com status 'Resigned' por departamento.
)

# Exibe o DataFrame resultante com as contagens.
dept_counts
```

```python
# Calcular taxa de atrito
# Divide o número de funcionários demitidos pelo total de funcionários por departamento e multiplica por 100 para obter a porcentagem.
dept_counts['attrition_rate_%'] = (dept_counts['resigned'] / dept_counts['total_emp']) * 100
```

```python
dept_counts
```

```python
# Classificar por taxa de atrito
# Ordena o DataFrame pela coluna 'attrition_rate_%' em ordem decrescente para identificar os departamentos com maior rotatividade.
dept_counts.sort_values("attrition_rate_%", ascending = False)
```
